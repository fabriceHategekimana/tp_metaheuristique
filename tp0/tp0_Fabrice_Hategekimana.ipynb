````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
````

## Simulation of a balanced dice


````python
def balanced_dice(N):
    return lambda : int(random.random()*N)
````

````python
def my_hist(generator, exp=100000):
    x = np.array([generator() for y in range(exp)])
    labels = np.unique(x)
    val, bins = np.histogram(x, len(labels))
    plt.bar(labels, val/float(len(x)), edgecolor="k", width=(max(labels)-min(labels))/float(len(labels)))
    plt.xlabel("labels")
    plt.ylabel("probabilities")
````

````python
my_hist(balanced_dice(6), 10000)
````

#### How would the histogram change if the sides are not equi-probable ?
If the side are not equi-probalbe, the bar aren't the same size (see roulette, methode below).

## Simulation of a biased coin toss

````python
def biased_coin_toss(p):
    return lambda : int(random.random()+p) 
````

````python
my_hist(biased_coin_toss(0.5))
````

````python
my_hist(biased_coin_toss(0.7))
````

#### What are the possible values of λ, and with what probabilities are they obtained ?

| lambda    | 1 | 0   |
|-----------|---|-----|
| P(lambda) | p | 1-p |



## Simulation of a double biased coin

To do so, we can use the previous function and label the result in this fashion.

```
[0,0] => 1*0 + 2*0 = 0
[1,0] => 1*1 + 2*0 = 1
[0,1] => 1*0 + 2*1 = 2
[1,1] => 1*1 + 2*1 = 3
```


````python
def double_biased_coin_toss(p1, p2):
    return lambda : biased_coin_toss(p1)()*1 + biased_coin_toss(p2)()*2
````

````python
my_hist(double_biased_coin_toss(0.5, 0.5))
````

````python
my_hist(double_biased_coin_toss(0.25, 0.5))
````

````python
my_hist(double_biased_coin_toss(0.25, 0.75))
````

## Roulette method

````python
def roulette_method(proba_list):
    return lambda : list(filter(lambda x: x >= random.random(), np.cumsum(proba_list)))[0]
````

````python
my_hist(roulette_method([0.25, 0.25, 0.25, 0.25]), 10000000)
````
